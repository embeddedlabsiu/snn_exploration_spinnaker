```python
% matplotlib inline

import matplotlib.pyplot as plt
```

```python
plotproperties = {'font.size': 12,
                  'axes.titlesize': 'xx-large',
                  'axes.labelsize': 'xx-large',
                  'xtick.labelsize': 'xx-large',
                  'xtick.major.size': 7,
                  'xtick.minor.size': 5,
                  'ytick.labelsize': 'xx-large',
                  'ytick.major.size': 7,
                  'ytick.minor.size': 5,
                  'legend.fontsize': 'large',
                  'figure.figsize': (7, 6),
                  'savefig.dpi': 300,
                  'savefig.format': 'jpg'}

import matplotlib as mpl
mpl.rcParams.update(plotproperties)
```

```python
ann_acc = 87.86
ann_err = 100 - ann_acc
values = [16.5, 59.82, 79.5, 83.6, 87.82]
values = [100 - v for v in values]
labels = ['default', '+ max normalization', '+ reset by subtraction', '+ analog input', '+ 99.9% normalization']
colors = ['blue', 'green', 'brown', 'orange', 'red']
fig, ax = plt.subplots()
for i in range(len(values)):
    b = ax.bar(i + 0.5, values[i], 0.8, label=labels[i], color=colors[i])[0]
    ax.text(b.get_x() + b.get_width()/2., 1.05*b.get_height(), '{:.2f}'.format(b.get_height()), ha='center', va='bottom',
            color=colors[i], fontsize=15, fontweight='bold')
ax.hlines(ann_err, 0, 5, 'black', 'dashed', label='ANN', linewidth=2)
ax.text(-0.7, 0.85*ann_err, str(ann_err), fontsize=15, fontweight='bold')
# ax.set_ylabel('Accuracy [%]')
ax.set_ylabel('Error [%]')
ax.set_ylim(0, 100)
ax.set_xlim(0, 5)
ax.set_xticks([])
ax.legend()
plt.savefig('/home/rbodo/Downloads/SNN', bbox_inches='tight')
```
