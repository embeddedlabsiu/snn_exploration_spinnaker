# Dataset

```python
import os
import h5py
import numpy as np
dataset_path = '/home/rbodo/.snntoolbox/Datasets/comma-driving'
with h5py.File(os.path.join(dataset_path, 'camera/') + '2016-01-30--11-24-51.h5', 'r') as f:
    data = np.array(f.get('X'))
data = data[1000:1100]
print(data.shape)
```

```python
x_test = (2. * (data / 255. - 0.5)).astype('float32')
y_test = np.zeros((len(x_test), 1), 'float32')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(x_test[0].transpose(1, 2, 0))
plt.show()
```

```python
np.savez_compressed(os.path.join(dataset_path, 'x_test'), x_test)
np.savez_compressed(os.path.join(dataset_path, 'y_test'), y_test)
```
